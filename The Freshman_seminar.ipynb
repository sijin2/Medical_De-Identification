```python
import pandas as pd
import numpy as np
import random
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
#피부과
dermatology = ["psoriasis", "Erythema Nodosum", "photosensitivity dermatitis",
              "mycoisis fungoides", "Sezary’s syndrome", "Hansen’s disease",
              "scrofuloderma", "Keloid scar", "dermatitis seborrheica", "fetid sweat"]

#안과
ophthalmology = ["corneal clouding", "keratohelcosis", "conjunctivitis",
                "scleritis", "scleromalacia", "refractive error", "glaucoma",
                "dacryocystitis", "persistence of pupillary membrane", "corectopia"]

#내과
internal = ["asthma", "diabetes", "constipation", "urticaria", "cystitis",
           "tetanus", "gout", "high blood pressure", "tuberculosis", "herpes zoster"]


#외과
surgery = ["facial paralysis", "intestinal obstruction", "varicose vein",
          "cholelithiasis", "anal fistula", "lipoma", "cholecystitis",
          "lung cancer", "stomach cancer", "pancreas cancer"]


#신경과
neurology = ["agnosia", "akinesia", "alert", "aphasia",
            "apraxia", "ataxia", "atrophy", "Aneursym",
            "brain death", "encephalitis"]


#정형외과
orthopedics = ["oblique fracture", "open fracture", "osteomyelitis", "arthritis",
              "spiral fracture", "pathologic fracture", "osteoporosis", "shaft fracture",
              "acute bronchitis", "transverse fracture"]

dis = []

dis = dermatology + ophthalmology + internal + surgery + neurology + orthopedics
len(dis)
```

```python
sr1 = pd.read_csv('Freshman.csv', sep=',', engine = 'python', encoding = "cp949")
sr1["Diseases"] = np.random.choice(dis, size = 1000)

sr1.to_csv("Ori_Freshman.csv", index = False)

sr2 = pd.DataFrame()

while(True):
    
    print("1. Masking Phone Number")
    print("2. Masking Name")
    print("3. Disease Categorization")
    print("4. Gender M-F Categorization")
    print("5. Gender F-M Categorization")
    print("6. Show Table")
    print("7. Reset")
    print("0. Exit")
    
    n1 = int(input("Choose the number you want.\n"))
    
    
    if(n1 == 1):
        
        sr1["Phone"] = "***-***-****"
        print("Masking Phone Number Success!")
        print()
        
    elif(n1 == 2):
        
        sr1["Name"] = "***"
        print("Masking Name Success!")
        print()
    
    elif(n1 == 3):
        
        sr2 = pd.DataFrame()
        
        print("1. dermatology")
        print("2. ophthalmology")
        print("3. internal")
        print("4. surgery")
        print("5. neurology")
        print("6. orthopedics")
        print("Use Comma and Except Gap")
        
        cls = input("Enter the department you want to see.").split(",")
        
        for i in range(len(cls)):
            
            sr3 = pd.DataFrame()
            
            if(cls[i] == '1'):
                
                for j in range(10):
                    
                    sr3 = pd.concat([sr3,sr1[sr1['Diseases'] == dermatology[j]]])
                
                sr3["Diseases"] = "dermatology"
                
                sr2 = pd.concat([sr2,sr3])
                    
                
            elif(cls[i] == '2'):
                
                for j in range(10):
                    
                    sr3 = pd.concat([sr3,sr1[sr1['Diseases'] == ophthalmology[j]]])
                
                sr3["Diseases"] = "ophthalmology"
                
                sr2 = pd.concat([sr2,sr3])
                
            elif(cls[i] == '3'):
                
                for j in range(10):
                    
                    sr3 = pd.concat([sr3,sr1[sr1['Diseases'] == internal[j]]])
                
                sr3["Diseases"] = "internal"
                
                sr2 = pd.concat([sr2,sr3])
                
            elif(cls[i] == '4'):
                
                for j in range(10):
                    
                    sr3 = pd.concat([sr3,sr1[sr1['Diseases'] == surgery[j]]])
                
                sr3["Diseases"] = "surgery"
                
                sr2 = pd.concat([sr2,sr3])
                
            elif(cls[i] == '5'):
                
                for j in range(10):
                    
                    sr3 = pd.concat([sr3,sr1[sr1['Diseases'] == neurology[j]]])
                
                sr3["Diseases"] = "neurology"
                
                sr2 = pd.concat([sr2,sr3])
                    
            elif(cls[i] == '6'):
                
                for j in range(10):
                    
                    sr3 = pd.concat([sr3,sr1[sr1['Diseases'] == orthopedics[j]]])
                
                sr3["Diseases"] = "orthopedics"
                
                sr2 = pd.concat([sr2,sr3])

                
            else:
                
                print("Wrong Number")
        
        
        sr2.to_csv("Fres_Used.csv", index = False)
        sr1 = pd.read_csv('Fres_Used.csv', sep=',', engine = 'python')
                
        print()
        
    elif(n1 == 4):
        
        sr1 = sr1.sort_values('Gender',ascending=False)
        print("Gender M-F Categorization Success!")
        print()
        
    elif(n1 == 5):
        
        sr1 = sr1.sort_values('Gender',ascending=True)
        print("Gender F-M Categorization Success!")
        print()
    
    elif(n1 == 6):
        
        print("Show Table!")
        print()
        sr1
        print()
        
    elif(n1 == 7):
        
        print("Reset!")
        sr1 = pd.read_csv('Ori_Freshman.csv', sep=',', engine = 'python')
        print()
        
    elif(n1 == 0):
        
        print("Exit!")
        print()
        sr1
        print()
        
        break
        
    
    
    else:
        print("Wrong Number.")
        print()
        


```
